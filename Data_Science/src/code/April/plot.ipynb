```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.close('all') 
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index, columns=list('ABCD'))
```

```python
df = df.cumsum()
```

```python
plt.figure();
```

```python
df.plot();
```
